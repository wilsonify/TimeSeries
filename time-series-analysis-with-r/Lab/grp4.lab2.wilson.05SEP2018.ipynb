1. Use the random numbers you generated in the previous lab and create a simple random walk process.

```r
w10 <- rnorm(10, 0, 1); x10 = cumsum(w10)
w30 <- rnorm(30, 0, 1); x30 = cumsum(w30)
w100 <- rnorm(100, 0, 1); x100 = cumsum(w30) 
w10000 <- rnorm(10000, 0, 1); x10000 = cumsum(w10000)
```

```r
plot(x10)
```

Estimate the ﬁrst 10 autocovariance and autocorrelation estimates, and produce a correlogram. 

```r
acf(x100, type = "covariance", lag.max = 10)
acf(x100, type = "correlation", lag.max = 10)
```

Does the ACF die if you increase the lag.max to its maximum value?

Describe your results. 

```r
acf(x100, lag.max = 100)
```

2. Generate an MA series $x_t = 0.6 w_t + 0.3 w_{t−1}$ and $n = 10,30,100,10000$. 

Produce correlograms and interpret.

```r
v10 = filter(w10, filter=c(0.6,0.3), method="convolution", sides = 1)
v30 =  filter(w30, filter=c(0.6,0.3), method="convolution", sides = 1)
v100 =  filter(w100, filter=c(0.6,0.3), method="convolution", sides = 1)
v10000 =  filter(w10000, filter=c(0.6,0.3), method="convolution", sides = 1)
```

```r
acf(v10)
acf(v30)
acf(v100)
acf(v10000)
```
